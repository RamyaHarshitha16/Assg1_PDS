```python
# eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
cleaned_data = pd.read_csv("/content/cleaned_data.csv")
# Descriptive statistics
print(cleaned_data.describe())

# Visualization 1: Histogram of Age
plt.figure(figsize=(8, 4))
sns.histplot(cleaned_data['Age'], bins=5, kde=True)
plt.title("Histogram for Age")
plt.show()

```

```python
# Visualization 2: Scatter plot of Grip strength vs. Age
plt.figure(figsize=(6, 6))
sns.scatterplot(x='Age', y='Grip strength (kg)', data=cleaned_data)
plt.title("Scatter plot of Grip strength vs. Age")
plt.show()
```
