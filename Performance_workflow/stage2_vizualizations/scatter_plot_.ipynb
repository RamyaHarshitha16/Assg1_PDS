```python

import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv('/content/cleaned_data.csv')
# Assuming your dataset has columns 'gender' and 'Average Score'
race = data['race/ethnicity']
average_scores = data['average score']
# Create a scatter plot
plt.figure(figsize=(8, 8))
plt.scatter(race, average_scores, alpha=0.5)
plt.xlabel('Race & Ethnicity')
plt.ylabel('Average Score')
plt.title('Scatter Plot of Race/Ethnicity v/s Average Score')

# Show the histogram
plt.show()
```
