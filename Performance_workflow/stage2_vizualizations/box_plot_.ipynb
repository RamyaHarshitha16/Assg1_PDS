```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv('/content/cleaned_data.csv')

# Assuming your dataset has columns 'parental level of education' and 'Average Score'
plt.figure(figsize=(12, 10))
sns.boxplot(x='parental level of education', y='average score', data=data, palette='Set1')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.title('Box Plot of Parental Level of Education v/s. Average Score')
plt.xticks(rotation=0, horizontalalignment='left')
# Show the histogram
plt.show()
```
