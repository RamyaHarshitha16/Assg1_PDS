```python
# Bar Chart of Gender vs. Average Score

import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv('/content/cleaned_data.csv')

# Assuming your dataset has columns 'race/ethnicity' and 'Average Score'
gender = data['gender']
average_scores = data['average score']

# Create a bar chart
plt.figure(figsize=(5, 5))
data_grouped = data.groupby('gender')['average score'].mean()
data_grouped.plot(kind='bar', color='red')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.title('Bar Chart of Gender vs. Average Score')
plt.xticks(rotation=45)
# Show the histogram
plt.show()
```
