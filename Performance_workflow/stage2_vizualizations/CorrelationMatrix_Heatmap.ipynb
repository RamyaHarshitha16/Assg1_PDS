```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv('/content/cleaned_data.csv')

# Assuming your dataset has columns 'lunch' and 'Average Score'
data['lunch'] = data['lunch'].map({'standard': 0, 'free/reduced': 1})

# Calculate the correlation matrix
correlation_matrix = data[['lunch', 'average score']].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap of Lunch v/s. Average_Score')
# Show the histogram
plt.show()
```
